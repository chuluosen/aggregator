```python
# 如果尚未安装 pandas
!pip install pandas

import os
import re
import pandas as pd

# 1) 指定你的数据库所在的根目录 (根据你的实际路径修改)
base_path = "/content/database"

# 2) 定义一个正则表达式，用于从文件名解析出 seed, age, sex, race
#    假设格式为: <seed>_<age>_<M/F>_<race>.png
pattern = re.compile(r'^(\d+)_(\d+)_(M|F)_(.+)\.png$', re.IGNORECASE)

all_records = []

# 3) 递归遍历 base_path 下所有子文件夹及文件
for root, dirs, files in os.walk(base_path):
    for filename in files:
        # 只关心 .png 文件，你也可以改成 .jpg 或其它
        if filename.lower().endswith(".png"):
            match = pattern.match(filename)
            if match:
                seed = match.group(1)       # 文件名中的 seed
                age = int(match.group(2))   # 文件名中的 age (转成 int)
                sex = match.group(3)        # M / F
                race = match.group(4)       # 余下的 race 信息 (如 indian、white、asian等)

                file_path = os.path.join(root, filename)  # 图像的完整路径

                # 4) 保存一条记录
                all_records.append({
                    "seed": seed,
                    "age": age,
                    "sex": sex,
                    "race": race,
                    "file_path": file_path
                })

# 5) 转为 DataFrame 方便后续处理
df = pd.DataFrame(all_records)
print(f"共收集到 {len(df)} 个文件记录。")

# 6) 查看前几行
df.head()
```

```python
import numpy as np

# 1) 先获取当前所有不重复的 seed
unique_seeds = df['seed'].unique()

# 2) 为了重复性，每次划分相同，可先设置随机种子
np.random.seed(42)  # 你可以换成其他数字

# 3) 将 unique_seeds 的顺序打乱
np.random.shuffle(unique_seeds)

# 4) 计算 train / val / test 的数量
n_total = len(unique_seeds)
train_size = int(0.7 * n_total)   # 70%
val_size   = int(0.1 * n_total)   # 10%
test_size  = n_total - train_size - val_size  # 剩余的 20%

# 5) 划分 ID
train_seeds = unique_seeds[:train_size]
val_seeds   = unique_seeds[train_size : train_size+val_size]
test_seeds  = unique_seeds[train_size+val_size :]

# 6) 建立 {seed -> split} 的字典映射
seed2split = {}
for s in train_seeds:
    seed2split[s] = "train"
for s in val_seeds:
    seed2split[s] = "val"
for s in test_seeds:
    seed2split[s] = "test"

# 7) 在原 DataFrame 中增加一列 "split" 表示所属集合
df['split'] = df['seed'].map(seed2split)

# 8) 检查统计结果
print("划分后，各 split 数量：")
print(df['split'].value_counts())

# 9) 也可以查看每个split中，seed的数量：
print("\n各 split 的种类（不同 seed）数：")
print(df.groupby('split')['seed'].nunique())

# 10) 查看前几行，确认列 "split" 已附加正确
df.head(10)
```

```python
class MyFaceDataset(Dataset):
    def __init__(self, df, transform=None):
        # 记录传入的 DataFrame，并重置索引(避免花式索引带来的混乱)
        self.df = df.reset_index(drop=True)
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, index):
        row = self.df.iloc[index]
        img_path = row['file_path']
        seed = row['seed']   # ID

        # 加载图像(PIL格式)
        image = Image.open(img_path).convert("RGB")

        # 应用预处理和数据增强
        if self.transform:
            image = self.transform(image)

        # 如果你想把 seed 也转为数值标签，可尝试:
        # seed = int(seed)  # 如果 seed 是字符串，就转为整数
        # 不过若是要做分类/embedding，还需建立 seed->label 的映射。这里先不展开。

        return image, seed

# 统一的 transform
# 下面示例把图像缩放到 112x112，再转为 [C,H,W] 范围[0,1] 的张量
common_transform = T.Compose([
    T.Resize((112, 112)),  # 或其他尺寸，比如 (128,128)
    T.ToTensor(),
])
```

```python
# 假设你的 df 已经包含 'split' 列
df_train = df[df['split'] == "train"]
df_val   = df[df['split'] == "val"]
df_test  = df[df['split'] == "test"]

train_dataset = MyFaceDataset(df_train, transform=common_transform)
val_dataset   = MyFaceDataset(df_val, transform=common_transform)
test_dataset  = MyFaceDataset(df_test, transform=common_transform)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader   = DataLoader(val_dataset, batch_size=32, shuffle=False)
test_loader  = DataLoader(test_dataset, batch_size=32, shuffle=False)
```
